```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
from matplotlib import cm
```

```python
"""
绘制折线图
"""
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 4, 6, 7]
y2 = [4, 3, 2, 5, 4]
# 绘制折线图
plt.plot(x, y1,color='red',label='A',marker='x',linewidth=2)
plt.plot(x, y2,color='blue',label='B',marker='o',linewidth=2,linestyle='--')
# 添加基本图形属性
plt.title('line chart')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend()
plt.ylim(0,10)
plt.show()
```

```python
"""
面积图
"""
x = [1, 2, 3, 4, 5]
y1=[80, 60, 80, 90, 90]
y2=[30, 45, 21 ,43, 56]
# 绘制面积图
plt.stackplot(x, y1, colors=['tan'],labels=['A'])
plt.stackplot(x, y2, colors=['red'],labels=['B'])
# 添加基本图形属性

plt.title('不知道')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend(loc=2)
plt.show()
```

```python
# 准备数据
x = [1, 2, 3, 4, 5]
y1=[80, 60, 80, 90, 90]
y2=[30, 45, 21 ,43, 56]
# 绘制堆积面积图
plt.subplots(figsize=(4.5,3.5),dpi=100,facecolor="blue")
plt.stackplot(x, [y1,y2], colors=['purple','red'],labels=['A','B'])
# 添加基本图形属性
plt.title('Stacked Area plot')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend(loc=2)
plt.show()
```

```python
"""
百分比堆叠图
"""
# 准备数据
x = [1, 2, 3, 4, 5]
y1 = np.array([2, 5, 4, 6, 7])
y2 = np.array([4, 3, 2, 5, 4])
y3 = np.array([8, 6, 8, 3, 3])
y1p = y1/(y1 + y2 + y3)*100
y2p = y2/(y1 + y2 + y3)*100
y3p = y3/(y1 + y2 + y3)*100
# 绘制百分比堆积面积图
plt.stackplot(x, [y1p, y2p,y3p],colors=['bisque','darkorange','tan'],labels=['A','B','C'])
# 添加基本图形属性
plt.title('难吃')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend()
plt.ylim(0,3)
plt.xlim(1,5)
plt.yticks(range(0,120,20),[f"{i}%" for i in range(0,120,20)])
plt.show()
```

```python
"""
四宫格
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
from matplotlib import cm
iris=pd.read_csv('iris.csv')
plt.rcParams['font.family'] = 'SimSun'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({
    'figure.facecolor': '#f8fbff',   # 淡蓝色背景
    'axes.facecolor': '#f0f8ff',     # 更淡的蓝色
    'axes.edgecolor': '#4a7d9a',     # 柔和的蓝灰色边框
    'grid.color': '#d6e7ff',         # 非常淡的蓝色网格
    'axes.labelcolor': '#1a3d5c',    # 深蓝色标签
    'xtick.color': '#2c5a78',        # 深蓝色刻度
    'ytick.color': '#2c5a78',
    'text.color': '#1a3d5c',         # 主要文本颜色
})

# 修改整个Figure背景为浅灰色（RGBA值）
plt.figure(figsize=(16,12), dpi=200)  # 浅灰色，不透明

# 统一的子图背景色设置（比Figure深）
subplot_bg = (0.88, 0.88, 0.88, 0.7)  # RGB 0.88比0.94深，透明度0.7
text_color = '#333333'  # 更深的文本颜色增强可读性

# 子图1 - 双圆环
ax1 = plt.subplot(221)
#ax1.set_facecolor(subplot_bg)  # 使用统一的深色背景
vals = np.array([[60, 32], [37, 40], [29, 10]])
cmap = plt.cm.Set2
outer_colors = cmap(np.arange(3) * 4)
inner_colors = cmap([1, 2, 5, 6, 9, 10])
plt.pie(vals.sum(axis=1),autopct='%1.1f%%', radius=1.2,pctdistance=0.83,colors=outer_colors,explode=(0, 0, 0.05),
        wedgeprops=dict(width=0.3, edgecolor='w'))
plt.pie(vals.flatten(),autopct='%1.1f%%', radius=1.2-0.3,pctdistance=0.78,colors=inner_colors,explode=(0, 0, 0, 0, 0, 0),
        wedgeprops=dict(width=0.3, edgecolor='w'),)
plt.title('双圆环', loc='center', fontsize=20, color=text_color)  # 缩小标题大小并使用深色

# 子图2 - 
ax2 = plt.subplot(222)
#ax2.set_facecolor(subplot_bg)  # 使用统一的深色背景
Iris=iris.drop('id',axis=1).drop('species',axis=1)
mask = np.triu(np.ones_like(Iris.corr(),dtype=bool),k=1)
seaborn.set_style("darkgrid", {"axes.facecolor": "#f0f0f0"})
seaborn.heatmap(Iris.corr(),mask=mask,cmap='Blues',
                center=0.4,
                linewidth=1.5,
                linecolor='white',
                square=True,
                annot=True,
                fmt='.2f',
                annot_kws={'size': 20, 'rotation': 45, 'ha': 'center', 'va': 'center'},
                cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient', 'orientation': 'vertical', 'ticks': [-1, -0.5, 0, 0.5, 1], 'drawedges': True},
                ax=ax2,
                )

# 子图3 - 折线图
plt.rcParams['font.family'] = 'SimSun'
plt.rcParams['axes.unicode_minus'] = False
ax3 = plt.subplot(223)
#ax3.set_facecolor(subplot_bg)  # 使用统一的深色背景
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_color('#666666')  # 添加深色边框线
ax3.spines['bottom'].set_color('#666666') # 添加深色边框线
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 4, 6, 7]
y2 = [4, 3, 2, 5, 4]

# 设置网格 - 调整颜色与背景协调
ax3.grid(True, which='major', 
         axis='both', 
         linestyle='--', 
         color='#aaaaaa',  # 比背景更深的网格线
         alpha=0.7, 
         linewidth=0.8)

# 绘制折线图 - 使用更饱和的颜色
plt.plot(x, y1, color='#e74c3c', label='A', marker='x', linewidth=2)  # 更鲜艳的红色
plt.plot(x, y2, color='#3498db', label='B', marker='o', linewidth=2, linestyle='--')  # 更鲜艳的蓝色
plt.title('折线图 - 带主刻度网格', fontsize=20, color=text_color)
plt.xlabel('x轴', fontsize=20, color=text_color)
plt.ylabel('y轴', fontsize=20, color=text_color)
plt.legend(loc='best', fontsize=20, facecolor=subplot_bg)  # 图例背景与子图协调
plt.ylim(0, 10)
# 添加白色背景的刻度标签更清晰
ax3.tick_params(axis='both', colors=text_color, which='both', labelsize=20)
for label in ax3.get_xticklabels() + ax3.get_yticklabels():
    label.set_backgroundcolor((1, 1, 1, 0.5))  # 半透明白色背景增强可读性

# 子图4 - 添加文本展示
ax4 = plt.subplot(224)
#ax4.set_facecolor(subplot_bg)  # 使用统一的深色背景
years = np.array(['2021', '2022', '2023', '2024'])  # 年份
x = np.arange(len(years))  # 年份对应的索引，用于设置柱的位置

# 随机生成四组数据
num_groups = 7  # 每组的数量
bar_width = 0.2  # 柱子的宽度
group_gap = 0.5  # 组与组之间的间距

# 计算每一组柱子的位置
# #计算每组柱子的宽度
group_width = num_groups * bar_width
# #计算每组柱子的起始位置
start_positions = np.arange(len(years)) * (group_width + group_gap) - (group_width / 2)


# 随机生成每组柱子的数据
data = np.random.randint(100, 700, size=(num_groups, len(years)))

# 画柱状图
for i in range(num_groups):
    plt.bar(start_positions + i * bar_width, data[i], width=bar_width, color=plt.cm.tab10(i), label=f'Data {i+1}', align='edge')

# 设置横轴标签和标题
plt.xlabel('Year', fontsize=25)
plt.ylabel('Value', fontsize=25)
plt.title('Bar Chart of Multiple Groups of Data for Each Year', fontsize=20)
plt.xticks(start_positions + group_width / 2, years)  # 设置横坐标为每组中心点

# 设置图例
plt.legend(loc='upper center',bbox_to_anchor=(1.1,1))
ax4.tick_params(axis='both', colors=text_color, which='both', labelsize=20)
# 增加子图间距以显示背景层次
plt.tight_layout(pad=3.0, w_pad=5.0, h_pad=5.0)  # 增大间距增强背景对比效果
plt.savefig('四宫格图.png', dpi=300, bbox_inches='tight')  # 保存图像
plt.show()
```
