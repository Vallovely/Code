```python
years = np.array(['2021', '2022', '2023', '2024','2025'])  # 年份
x = np.arange(len(years))  # 年份对应的索引，用于设置柱的位置

num_groups = 8  # 每组的数量
bar_width = 0.2  # 柱子的宽度
group_gap = 0.5  # 组与组之间的间距

# 计算每一组柱子的位置
# #计算每组柱子的宽度
group_width = num_groups * bar_width
# #计算每组柱子的起始位置
start_positions = np.arange(len(years)) * (group_width + group_gap) - (group_width / 2)


# 随机生成每组柱子的数据
data = np.random.randint(100, 700, size=(num_groups, len(years)))

# 画柱状图
for i in range(num_groups):
    plt.barh(start_positions + i * bar_width, data[i], height=bar_width, color=plt.cm.tab10(i), label=f'Data {i+1}', align='edge')

# 设置横轴标签和标题
plt.ylabel('Year')
plt.xlabel('Value')
plt.title('Bar Chart of Multiple Groups of Data for Each Year')
plt.yticks(start_positions + group_width / 2, years)  # 设置横坐标为每组中心点

# 设置图例
plt.legend()

# 显示图像
plt.show()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 设置全局字体
plt.rcParams['font.sans-serif'] = ['SimSun']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 创建示例数据
age_groups = ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', 
              '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', 
              '65-69', '70-74', '75-79', '80-84', '85+']

# 男性人口比例（%，使用负值使其在左侧显示）
male = [-5.2, -5.5, -5.8, -6.0, -6.5, -7.0, -6.8, -6.5, 
        -6.0, -5.5, -4.8, -4.2, -3.8, -3.2, -2.8, -2.0, -1.2, -0.8]

# 女性人口比例（%）
female = [5.0, 5.3, 5.6, 5.9, 6.3, 6.8, 6.6, 6.3, 
          5.9, 5.4, 4.9, 4.5, 4.2, 3.8, 3.4, 2.7, 2.0, 1.5]

# 创建画布和子图
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
fig.patch.set_facecolor('#f8f9fa')  # 设置画布背景色

# 绘制男性人口条形图（左侧）
male_bars = ax.barh(age_groups, male, color='#1f77b4', label='男性', edgecolor='white', linewidth=0.7)

# 绘制女性人口条形图（右侧）
female_bars = ax.barh(age_groups, female, color='#e377c2', label='女性', edgecolor='white', linewidth=0.7)

# 设置图表标题和轴标签
ax.set_title('2023年中国人口年龄金字塔图', fontsize=18, pad=20, fontweight='bold', color='#2a3f5f')
ax.set_xlabel('人口比例 (%)', fontsize=12, labelpad=15, color='#555555')
ax.set_ylabel('年龄组', fontsize=12, labelpad=15, color='#555555')

# 设置x轴刻度和标签
max_value = max(max(np.abs(male)), max(female))
ax.set_xlim(-max_value*1.1, max_value*1.1)
ax.set_xticks(np.arange(-8, 9, 2))
ax.set_xticklabels([str(abs(x)) for x in np.arange(8, -9, -2)])  # 使两侧标签都显示为正数

# 添加参考线
ax.axvline(0, color='gray', alpha=0.5)
for x in np.arange(-8, 9, 2):
    ax.axvline(x, color='gray', alpha=0.1, linestyle='--')

# 添加数据标签
for bar in male_bars:
    width = bar.get_width()
    if width < -0.5:  # 只对较大的条形添加标签
        ax.text(width + 0.1, bar.get_y() + bar.get_height()/2, 
                f'{abs(width):.1f}', 
                ha='right', va='center', color='white', fontsize=8)

for bar in female_bars:
    width = bar.get_width()
    if width > 0.5:
        ax.text(width - 0.1, bar.get_y() + bar.get_height()/2, 
                f'{abs(width):.1f}', 
                ha='left', va='center', color='white', fontsize=8)

# 添加图例
ax.legend(loc='upper right', ncol=2, bbox_to_anchor=(0.9,1.0), 
          frameon=True, framealpha=0.9)

# 添加注释说明
plt.figtext(0.5, 0.01, '数据来源：中国国家统计局（示例数据）', 
            ha='center', fontsize=10, color='#777777')

# 调整布局
plt.tight_layout()
plt.subplots_adjust(bottom=0.1)

# 添加网格线（在条形后面）
ax.grid(axis='x', linestyle='--', alpha=0.3, color='gray')
ax.set_axisbelow(True)

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 生成数据
y=[1,2,3]
x1=np.array([585,532,471])
x2=np.array([320,405,461])
x3=np.array([76,97,84])
# 计算百分比
x1p = x1/(x1 + x2 + x3)*100
x2p = x2/(x1 + x2 + x3)*100
x3p = x3/(x1 + x2 + x3)*100
# 堆积百分比条形图
fig,ax = plt.subplots(figsize=(4.5,3.5),dpi=100,facecolor="w")
a=plt.barh(y, x1p, height=0.5,color='r', label='A')
b=plt.barh(y, x2p, height=0.5,left=x1p, color='g', label='B')
c=plt.barh(y, x3p, height=0.5,left=x1p+x2p, color='c', label='C')

# 显示范围，设置x轴区间为百分比。f“{}” 表示精准匹配
plt.yticks(y,[2017,2018,2019])
plt.xticks(np.arange(0,120,20),[f"{i}%" for i in range(0,120,20)])
# 添加图例
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 生成数据
x = np.linspace(0, 20, 20)
y1 = np.random.randint(50, 100, 20)
y2 = np.random.randint(50, 100, 20)
y3 = np.random.randint(50, 100, 20)
# 计算百分比
y1p = y1/(y1+y2+y3)*100
y2p = y2/(y1+y2+y3)*100
y3p = y3/(y1+y2+y3)*100
#带标签的百分比堆积柱状图（竖）
plt.rcParams['font.sans-serif']=['Simhei']
a=plt.barh(x, y1p, color='r', label='语文')
b=plt.barh(x, y2p, bottom=y1p, color='g', label='数学')
c=plt.barh(x, y3p,bottom=y1p+y2p, color='c', label='英语')
plt.barh_label(a,label_type='center',fmt='%d%%')
plt.barh_label(b,label_type='center',fmt='%d%%')
plt.bar_label(c,label_type='center',fmt='%d%%')

# 显示范围，设置纵轴区间为百分比。f“{}” 占位符表示精准匹配
plt.xlim(-2, 26)
plt.yticks(np.arange(0,120,20),[f"{i}%" for i in range(0,120,20)])
# 添加图例
plt.legend(loc='upper right')
plt.xlabel('学生人数', fontsize=15)
plt.ylabel('各科成绩占比', fontsize=15)
plt.grid(axis='y', color='gray', linestyle=':', linewidth=2)
plt.show()
```

```python
#导入所需要的功能
import pandas
iris=pandas.read_csv('iris.csv')
import matplotlib.pyplot as plt
import numpy as np              

 #定义三维数据数组
x = np.array(iris['petal_width']) 
y = np.array(iris['petal_length'])
z = np.array(iris['sepal_length'])

#建立画三维图方法
fig = plt.figure(figsize=(16,15))     #创建自定义的图像，（）为一切设置（背景 颜色、图像尺寸大小等）
ax=plt.subplot(projection='3d') #指定画的是3D图

#按要求画出三维散点图
ax.scatter(x[:50], y[:50], z[:50], c='r')      
ax.scatter(x[50:100], y[50:100], z[50:100], c='m')  
ax.scatter(x[100:], y[100:], z[100:], c='b')

 # 设置X、Y、Z轴的名字显示，用刺眼的红色
ax.set_xlabel('petal_width', fontdict={'size': 15, 'color': 'red'}) 
ax.set_ylabel('petal_length', fontdict={'size': 15, 'color': 'red'})
ax.set_zlabel('sepal_length', fontdict={'size': 15, 'color': 'red'})         


plt.show() 
```
